```python
# importa e carrega dados
import pandas as pd
% matplotlib inline

df_powerplant = pd.read_csv('powerplant_data_edited.csv')
df_powerplant.info()
```

```python
# trace um gráfico com a relação entre temperatura e saída elétrica
# df_powerplant.head()
df_powerplant.plot(x='Temperature', y='Electrical energy output', kind='scatter')
```

```python
# trace um gráfico com a distribuição da umidade
df_powerplant['Relative Humidity'].plot(kind='hist')
```

```python
# trace um gráfico de caixas para cada variável
pd.plotting.scatter_matrix(df_powerplant, figsize=(8,8));
```
